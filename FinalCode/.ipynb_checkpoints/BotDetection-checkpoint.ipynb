### Using Decision Tree Classifier

```python
#from sklearn.model_selection import train_test_split
#train_df, test_df = train_test_split(df, test_size=0.3)

filepath = 'https://raw.githubusercontent.com/jubins/ML-TwitterBotDetection/master/FinalCode/kaggle_data/'

training_data = pd.read_csv(filepath+'training_data_2_csv_UTF.csv')
```

```python
train_bots = training_data[training_data.bot==1]
train_nonbots = training_data[training_data.bot==0]
bots['screen_name_binary'] = (bots.screen_name.str.contains("bot", case=False)==True)

nonbots['screen_name_binary'] = (nonbots.screen_name.str.contains("bot", case=False))

train_df = pd.concat([bots, nonbots])

test_df = pd.read_csv(filepath+'test_data_4_students.csv', sep='\t')

test_df['screen_name_binary'] = (test_df.screen_name.str.contains("bot", case=False)==True)

train_df = df
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

clf = DecisionTreeClassifier(criterion='entropy')

X_train = train_df[['screen_name_binary']] #train_data 
y_train = train_df['bot'] #train_target

X_test = test_df[['screen_name_binary']] #test_data

#Training on decision tree classifier
model = clf.fit(X_train, y_train)

#Predicting on test data
predicted = model.predict(X_test)

test_df['bot'] = predicted
test_df['id'] = test_df['id'].apply(lambda x: int(x))

test_df = test_df[['id', 'bot']]
test_df.head(1)
```

```python
test_df.to_csv('submission.csv', index=False)
```
